```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
users = pd.read_csv('Users.csv')
```

```python
users.isna().sum()
```

```python
users.info()
```

```python
sns.kdeplot(data = users ,  x = 'Age')
```

```python
#Observayions..
#some age > 100 are given - [invalid age]
# its forms left skewed data 

"""""
solution :
!) we will delete the data with higher age as they are very less in number

""""

```

```python
users = users[users["Age"] < 101]

```

```python
sns.kdeplot(data = users ,  x = 'Age')
```

```python
#NOw as the data is slightly left skewed we will replace the null values for age with median

users['Age'].fillna(users['Age'].median() , inplace = True)  
```
