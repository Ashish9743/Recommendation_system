```python
import pandas as pd
import numpy as np
```

```python

books = pd.read_csv('Books.csv')
ratings = pd.read_csv('Ratings.csv')

```

```python
merged_data = pd.merge(ratings, books, on='ISBN')
merged_data.head()
```

```python
top_rating = merged_data['Book-Rating'].max()
top_books = merged_data[merged_data['Book-Rating'] == top_rating]
top_books.head()
```

```python
ratings = int(input('Enter ratings : '))
top_books = merged_data[merged_data['Book-Rating'] == ratings]
top_books.head()
```
